### Process to Extract Data

```python
files = ['refugees to Kenya per sex','refugees to Kenya per location', 'refugees to DRC per DRC region', 'refugees to Ethiopia per Ethiopian district','refugees to Sudan per Sudanese state','refugees to Uganda per Ugandan district']

for file in files:
    with open(file + '.json') as f:
        data = json.load(f)
        
    new_df = pd.DataFrame(data['data'])
    new_df.to_csv(file + "_cleaned.csv", sep=',')
```

#### Extracted Data

```python
from IPython.display import display
for file in files:
    df = pd.read_csv(file + '_cleaned.csv')
    print("#File: " + file)
    display(df.head())
```
